```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Eph = [164463, 399740, 654596, 7180, 8164]
Eph_err = [948, 3103, 6672, 171, 213]
E = [662, 1592, 2600, 29.7, 34]
```

```python
plt.figure(figsize=[10,5])
plt.errorbar(E, Eph, yerr = Eph_err, fmt='o');

fit =np.polyfit(E, Eph, 1)
m, c = fit

Ew =np.linspace(np.min(E), np.max(E), 50)
plt.plot(Ew, m*Ew+c)

plt.xlabel('Energy (keV)');
plt.ylabel('Energy (pes)');
plt.title('Calibration plot');
```
